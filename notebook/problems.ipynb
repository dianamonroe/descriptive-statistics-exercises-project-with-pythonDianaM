# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
import matplotlib.pyplot as plt

#1.1 Generate an array of 100 elements following a normal distribution.


normal = np.random.normal(size = 100)
print("Normal vector: \n", normal)


#1.2 Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.

chi = np.random.chisquare(3, 100)
print("\n Chi vector: \n", chi)


plt.hist(normal, label = "Normal distribution")
plt.hist(chi, label = "Chi distribution", alpha=0.5)
plt.legend()
````

````python
#1.3) Calculate the main metrics and statistical measures that best describe the two vectors.

import statistics as stats

np.random.seed(42)

#MEAN
print("Numpy Normal mean: ", round(np.mean(normal), 4))
print("Stats Normal mean: ", round(stats.mean(normal), 4))

print("Numpy Chi mean: ", round(np.mean(chi), 4))
print("Stats Chi mean: ", round(stats.mean(chi), 4))


#MEDIAN
print("Numpy Normal median: ", round(np.median(normal), 4))
print("Stats Normal median: ", round(stats.median(normal), 4))

print("Numpy Chi median: ", round(np.median(chi), 4))
print("Stats Chi median: ", round(stats.median(chi), 4))


#MODE
print("Stats Normal mode: ", round(stats.mode(normal), 4))
print("CStats hi mode: ",round(stats.mode(chi), 4))


#RANGE
normal_range = max(normal) - min(normal)
print("Normal range is: ", round(normal_range, 4))

chi_range = max(chi) - min(chi)
print("Chi range is: ", round(chi_range, 4))


#VARIANCE
normal_var = stats.variance(normal)
print("Stats Normal variance is: ", round(normal_var, 4))
normal_var_np = np.var(normal)
print("Numpy Normal variance is: ", round(normal_var_np, 4))

chi_var = stats.variance(chi)
print("Stats Chi variance is: ", round(chi_var, 4))
chi_std_np = np.var(chi)
print("Numpy Chi variance is: ", round(chi_std_np, 4))


#STANDARD DEVIATION
normal_stdev = stats.stdev(normal)
print("Stats Normal Standard Dev.:", round(normal_stdev, 4))
normal_std_np = np.std(normal)
print("Numpy Normal Standard Dev.: ", round(normal_std_np, 4))

chi_stdev = stats.stdev(chi)
print("Stats Chi Standard Dev.: ", round(chi_stdev, 4))
chi_std_np = np.std(chi)
print("Numpy Chi Standard Dev.. ", round(chi_std_np, 4))


#SKEWNESS
  #Using scipy.stats
from scipy.stats import skew
print("Scipy.stats Normal Skew: ", round(skew(normal), 4))
print("Scipy.stats Chi Skew: ", round(skew(chi), 4))


  #USING Pandas
import pandas as pd
dfnormal = pd.Series(normal)
dfchi = pd.Series(chi)

print("Pandas Normal skew: ", round(dfnormal.skew(), 4))
print("Pandas Chi skew: ", round(dfchi.skew(), 4))

#KURTOSIS
from scipy.stats import kurtosis

print("Scipy.stats Normal Kurtosis: ", round(kurtosis(normal), 4))
print("Pandas Chi Kurstosis: ", round(dfchi.kurtosis(), 4))

print("Scipy.stats Chi Kurtosis: ", round(kurtosis(chi), 4))
print("Pandas Normal Kurtosis: ", round(dfnormal.kurtosis(), 4))


#Statistics overiew using Pandas.describe():
pandas_stats_normal = dfnormal.describe()
pandas_stats_chi = dfchi.describe()

print("\n Pandas Normal stats overview:\n ", pandas_stats_normal)
print("\n Pandas Chi stats overview: \n", pandas_stats_chi)
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
import numpy as np
import statistics as stats

data = [4, 2, 5, 8, 6]

media = sum(data) / len(data)
print(media)

def variance(data):
    return sum((i - media)**2 for i in data) / (len(data)-1)

standard_deviation = np.sqrt(variance(data))

print("Hand-made standard deviation: ", standard_deviation)

print("Numpy standard deviation: ", np.std(data))

print("Stats standard deviation: ", stats.stdev(data))
````
